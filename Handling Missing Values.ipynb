# Pandas has a better functionality that it consider missing value to Nan.

```python
import pandas as pd
```

```python
df=pd.read_csv('E:\Records2.csv')
df
```

## na_values parameter: It convert any missing , invalid value to Nan.

```python
df=pd.read_csv('E:\Records2.csv')
df
```

```python
df=pd.read_csv('E:\Records2.csv',na_values='Invalid')
df
```

```python
df=pd.read_csv('E:\Records2.csv',keep_default_na=False)
df
```

## isnull() method: It helps to check the missing and invalid values in the Data Frame.

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df.isnull().sum().sum()
```

## notnull() method: This method is just opposite of isnull() method.

```python
df.notnull()
```

```python
df.notnull().sum()
```

```python
df.notnull().sum().sum()
```

```python
df
```

## dropna() method: It by default remove those row or columns which has null or missing values.

```python
df.dropna()
```

```python
df.dropna(axis=1)
```
