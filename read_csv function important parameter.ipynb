# read_csv important attributes

```python
from pandas import *
```

```python
df1=read_csv('E:\Records2.csv')
df1
```

## List of all attributes in data frame

```python
print(df1.columns)
```

## nrows parameter: It is used to select rows from csv file

```python
df1=read_csv('E:\Records2.csv',nrows=6)
df1
```

```python
df1=read_csv('E:\Records2.csv',nrows=11)
df1
```

## usecols parameter: It print any specific column.

```python
df1=read_csv('E:\Records2.csv',usecols=[0])
df1
```

```python
df1=read_csv('E:\Records2.csv',usecols=[1,2])
df1
```

```python
type(df1)
```

## skiprows: It skips specific rows from the data frame.

```python
df1=read_csv('E:\Records2.csv',skiprows=[0,1,2])
df1
```

## index_col: It converts any column of Data Frame into its index column.

```python
df1=read_csv('E:\Records2.csv',index_col='Region')
df1
```

```python
df1=read_csv('E:\Records2.csv',index_col='Country')
df1
```

```python
df1=read_csv('E:\Records2.csv',index_col='Order ID')
df1
```

## header: It creates any particular row to header or heading.

```python
df1=read_csv('E:\Records2.csv',header=1)
df1
```

```python
df1=read_csv('E:\Records2.csv',header=10)
df1
```

## prefix: It create header of strings.

```python
df1=read_csv('E:\Records2.csv',header=None,prefix='Column_')
df1
```

## names: It helps to given particular specific name to each column.

```python
df1=read_csv('E:\Records2.csv',names=['Id','R'])
df1
```

## dtype: It can convert the data type of any column.

```python
df1=read_csv('E:\Records2.csv')
df1
```

```python
df1=read_csv('E:\Records2.csv',dtype={'Units sold':'float64'})
df1
```
