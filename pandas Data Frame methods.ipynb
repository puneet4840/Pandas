# Here are two methos to retrieve upper or lower data from Data Frame.

```python
from pandas import *
```

```python
df=read_csv('E:\Records2.csv')
df
```

## head() method: It by default returns the 5 upper rows from data frame.

```python
df=read_csv('E:\Records2.csv')
df.head()
```

```python
df=read_csv('E:\Records2.csv')
df.head(2)
```

## tail() method: It by default returns the 5 lower rows from data frame.

```python
df=read_csv('E:\Records2.csv')
df.tail()
```

```python
df=read_csv('E:\Records2.csv')
df.tail(2)
```
