# Three Methods are there to fill values in pandas Data Frame

## 1: fillna() method: It fill Nan values to given values.
## 2: interpolate() method: It first guess the values and then fills the correct guessed values in          place of Nan or missing values (integer values only).

```python
from pandas import *
```

```python
df=read_csv("E:\Records2.csv")
df
```

## fillna() method

```python
df.fillna(100,inplace=True)
df
```

## interpolate() method

```python
df.interpolate(inplace=True)
df
```
